```python
df = pd.read_csv('carprices.csv')
df
```

converting car names to values using pandas get dummies

```python
dummy = pd.get_dummies(df.Car_Model)
dummy
```

merging the two datas  using concatenate with pandas

```python
merged = pd.concat([df,dummy],axis = 'columns')
merged
```

```python
n_df = merged.drop(['Car_Model','Audi_A5'], axis = 'columns')
n_df
```

```python
x =n_df.drop(['Sell_Price'], axis = 'columns')
y = n_df.Sell_Price
```

```python
ln_model = LinearRegression()
ln_model.fit(x,y)
```
